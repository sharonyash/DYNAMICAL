# EULER METHOD

```python
# non linear ODE of order 1
# x-dot =  dx/dt = f(x)
# Euler method of this ODE : x_n+1 = x_n + f(x_n) * delta(t)
# Delta(t) = dt : time step 
# Time t = n*dt.
# r : control parameter. I set r = 1.0
# Our ODE is: Pitchfork Bifurcation Form
# x_dot = r * x - x * x * x
#r in [-1,1] is the control parameter

# IMPORT
from pylab import *
     
# DEFINE THE FUNCTION AND EULER METHOD
def ODEfunction(r,x):
    return r*x - x*x*x

def Euler1(r,x,f,dt):
    return x +  dt*f(r,x)
#return x +  (r*x - x*x*x*x) - this was wrong 

# Number of (dt)s to integrate 
N = 1000

# TIME STEP 
dt = 0.001 
r = 1.0
# TIME
t  = [ 0.0]

# INITIAL CONDITIONS for iterations of each x
x1 = [ 1]
x2 = [ -1]
x3 = [ 2.5]
x4 = [-2.5]
x5 = [ 3.5]
x6 = [-3.5]

# Iterate first, then calculate new x at each (dt) and we do x_n.append to add to the list
for n in range(0,N):
    x1.append( Euler1(r,x1[n],ODEfunction,dt) )
    x2.append( Euler1(r,x2[n],ODEfunction,dt) )
    x3.append( Euler1(r,x3[n],ODEfunction,dt) )
    x4.append( Euler1(r,x4[n],ODEfunction,dt) )
    x5.append( Euler1(r,x5[n],ODEfunction,dt) )
    x6.append( Euler1(r,x6[n],ODEfunction,dt) )
    t.append(t[n] + dt)

# PLOT
xlabel('t') # set x-axis label
ylabel('x') # set y-axis label
title('Euler Method for various Initial Condition at r= ' + str(r)) # set plot title
axis([0.0,dt*(N+1),-4.0,4.0])
# Plot the time series
plot(t,x1,'b', label =  'x1 = 1')
plot(t,x2,'r', label =  'x2 = -1')
plot(t,x3,'g', label =  'x3 = 2.5')
plot(t,x4,'m', label =  'x4 = -2.5')
plot(t,x5,'c', label =  'x5 = 3.5')
plot(t,x6,'y', label =  'x6 = -3.5')
#plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
show()
```

# Runga Kutta

```python
# non linear ODE of order 1
# x-dot =  dx/dt = f(x)
# Euler method of this ODE : x_n+1 = x_n + f(x_n) * delta(t)
# Delta(t) = dt : time step 
# Time t = n*dt.
# r : control parameter. I set r = 1.0
# Our ODE is: Pitchfork Bifurcation Form
# x_dot = r * x - x * x * x
#r in [-1,1] is the control parameter

# IMPORT
from pylab import *
 
    
# DEFINE THE FUNCTION AND EULER METHOD
def ODEfunction(r,x):
    return r*x - x*x*x

# Runge-Kutta Euler Method 4th order
def RungaKutta(r,x,f,dt):
    k1 = dt * f(r,x)
    k2 = dt * f(r,x + k1/2.0)
    k3 = dt * f(r,x + k2/2.0)
    k4 = dt * f(r,x + k3)
    return x + ( k1 + 2.0 * k2 + 2.0 * k3 + k4 ) / 6.0

# Number of (dt)s to integrate 
N = 1000

# TIME STEP 
dt = 0.001 
r = 2.1

# TIME
t  = [ 0.0]

# INITIAL CONDITIONS for iterations of each x
x11 = [0.2]
x22 = [-0.2]
x1 = [ 1.0]
x2 = [-1.0]
x3 = [ 2.5]
x4 = [-2.5]
x5 = [ 3.5]
x6 = [-3.5]

# Iterate first, then calculate new x at each (dt) and we do x_n.append to add to the list
for n in range(0,N):
    x1.append( RungaKutta(r,x1[n],ODEfunction,dt) )
    x2.append( RungaKutta(r,x2[n],ODEfunction,dt) )
    x3.append( RungaKutta(r,x3[n],ODEfunction,dt) )
    x4.append( RungaKutta(r,x4[n],ODEfunction,dt) )
    x5.append( RungaKutta(r,x5[n],ODEfunction,dt) )
    x6.append( RungaKutta(r,x6[n],ODEfunction,dt) )
    x11.append( RungaKutta(r,x11[n],ODEfunction,dt) )
    x22.append( RungaKutta(r,x22[n],ODEfunction,dt) )
    t.append(t[n] + dt)

# PLOT
xlabel('t, where dt = '  + str(dt)) # set x-axis label
ylabel('x') # set y-axis label
title('RungaKutta Method for various Initial Condition at r= ' + str(r)) # set plot title
axis([0.0,dt*(N+1),-4.0,4.0])
# Plot the time series
plot(t,x11,'c', label =  'x11 = 0.2')
plot(t,x22,'y', label =  'x22 = -0.2')
plot(t,x1,'b', label =  'x1 = 1')
plot(t,x2,'r', label =  'x2 = -1')
plot(t,x3,'g', label =  'x3 = 2.5')
plot(t,x4,'m', label =  'x4 = -2.5')
plot(t,x5,'c', label =  'x5 = 3.5')
plot(t,x6,'y', label =  'x6 = -3.5')

plt.legend()

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```
